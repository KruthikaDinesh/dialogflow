```
import pandas as pd
```

```
import matplotlib.pyplot as plt
```

```
data = pd.read_csv('NYPD_Complaint_Data_Historic.csv')
```

```
data.plot(kind="scatter", x="Longitude", y="Latitude", alpha=0.4)
plt.show()
```

```
import matplotlib.image as mpimg
ny_img=mpimg.imread('nyc-boroughs-map.jpg')
ax = data.plot(kind="scatter", x="Longitude", y="Latitude", alpha=0.4)
plt.imshow(ny_img, extent=[-124.55, -113.80, 32.45, 42.05], alpha=0.5)
plt.ylabel("Latitude", fontsize=14)
plt.xlabel("Longitude", fontsize=14)
plt.show()
```
